# Line plot Visualization

```python
New_Dataset1 = New_Dataset[New_Dataset['Timestamp'] > (New_Dataset['Timestamp'].max()-650000)]  
New_Dataset1.shape 
```

```python
New_Dataset1.head(2)
```

```python
New_Dataset1 = New_Dataset1.reset_index(drop = True) 
New_Dataset1.head(3) 
```

```python
# define figure size.
ploting.figure(figsize = (20,10)) 
ploting.rcParams.update({'font.size': 10}) 
ploting.subplot(2,1,1)
ploting.plot(New_Dataset1['Open'].head(1000))
ploting.legend(['Open price']) 
# x_label set here.
ploting.xlabel('Time period')
# y_label set here.
ploting.ylabel('Opening price')
# title set here.
ploting.title('Total number of Opening price of Bitcoin for Starting 1000 timestamps') 

ploting.subplot(2,1,2)
ploting.plot(New_Dataset1['Volume_(Currency)'].head(1000)) 
ploting.legend(['Volume currency'])   
ploting.xlabel('Time period')
ploting.ylabel('Volume Traded')
ploting.title('Total number of Volume traded of Bitcoin for Starting 1000 timestamps') 
ploting.show() 
```

# Bar graph Visualization

```python
New_Dataset['Timestamp'] = [datetime.fromtimestamp(x) for x in New_Dataset['Timestamp']] 
New_Dataset.head(2) 
```

```python
# Here we have extracting Date, year, month from our Timestamp column
New_Dataset['Date'] = New_Dataset['Timestamp'].dt.date
New_Dataset['year'] = New_Dataset['Timestamp'].dt.year 
New_Dataset['month'] = New_Dataset['Timestamp'].dt.month 
New_Dataset.head(2) 
```

```python
data_grouped1 = New_Dataset.groupby('year').mean() 
data_grouped1.head()
```

```python
df=data_grouped1
df['year']=df.index
df=df.reset_index(drop = True) 
df.head()
```

```python
df1=df[['year','Open','Close']] 
print(df1)

X = ["2011","2012","2013","2014","2015","2016","2017","2018","2019","2020","2021"]
X_axis=np.arange(len(X))  

df1.plot(x='year',kind='bar',stacked=False) 
ploting.xticks(X_axis, X)
# x_label set here.
ploting.xlabel("Years")
# y_label set here.
ploting.ylabel("Comparision") 
# title set here.
ploting.title("Comparision Of open & close Price") 
ploting.legend()
ploting.show() 
```
